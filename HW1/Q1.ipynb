```python
def matlab_style_gauss2D(shape=(3,3),sigma=0.5):
    """
    2D gaussian mask - should give the same result as MATLAB's
    fspecial('gaussian',[shape],[sigma])
    """
    m,n = [(ss-1.)/2. for ss in shape]
    y,x = np.ogrid[-m:m+1,-n:n+1]
    h = np.exp( -(x*x + y*y) / (2.*sigma*sigma) )
    h[ h < np.finfo(h.dtype).eps*h.max() ] = 0
    sumh = h.sum()
    if sumh != 0:
        h /= sumh
    return h
```

```python
def roundAngle(angle_array):
    resultArray = np.array([])
    for x in np.nditer(angle_array):
        if (-22.5 < x < 22.5 or x > 157.5 or x < -157.5):
            resultArray = np.append(resultArray, 0)
        elif (22.5 <= x <= 67.5 or -157.5 <= x <= -112.5):
            resultArray = np.append(resultArray, 45)
        elif (67.5 < x < 112.5 or -112.5 < x < -67.5):
            resultArray = np.append(resultArray, 90)
        else:
            resultArray = np.append(resultArray, 135)
    resultArray.shape = angle_array.shape
    return resultArray
        
```

```python
def supressNonMax(im, thet):
    result_im = np.array([])
    for i in range(0, im.shape[0]):
        for j in range(0, im.shape[1]):
            if (thet[i,j] == 0):
                if i == 0:
                    if im[i,j] > im[i+1,j]:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
                elif i == im.shape[0]-1:
                    if im[i,j] > im[i-1,j]:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
                else:
                    if (im[i,j] > im[i-1,j] and im[i,j] > im[i+1,j]):
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
            elif (thet[i,j] == 90):
                if j == 0:
                    if im[i,j] > im[i,j+1]:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
                elif j == im.shape[1]-1:
                    if im[i,j] > im[i,j-1]:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
                else:
                    if (im[i,j] > im[i,j+1] and im[i,j] > im[i,j-1]):
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
            elif (thet[i,j] == 45):
                if i == 0:
                    if j == 0:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i+1,j-1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if j == 0:
                    if i == 0:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i-1,j+1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if i == im.shape[0]-1:
                    if j == im.shape[1]-1:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i-1,j+1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if j == im.shape[1]-1:
                    if i == im.shape[0]-1:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i+1,j-1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                else:
                    if (im[i,j] > im[i-1,j+1] and im[i,j] > im[i+1,j-1]):
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
            elif (thet[i,j] == 135):
                if i == 0:
                    if j == im.shape[1]-1:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i+1,j+1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if j == 0:
                    if i == im.shape[0]-1:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i+1,j+1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if i == im.shape[0]-1:
                    if j == 0:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i-1,j-1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                if j == im.shape[1]-1:
                    if i == 0:
                        result_im = np.append(result_im, im[i,j])
                    else:
                        if im[i,j] > im[i-1,j-1]:
                            result_im = np.append(result_im, im[i,j])
                        else:
                            result_im = np.append(result_im, 0)
                    break
                else:
                    if (im[i,j] > im[i-1,j-1] and im[i,j] > im[i+1,j+1]):
                        result_im = np.append(result_im, im[i,j])
                    else:
                        result_im = np.append(result_im, 0)
                        
    #result_im.shape = im.shape
    return result_im
            
```
